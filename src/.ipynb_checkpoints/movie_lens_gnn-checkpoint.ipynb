# 1. Setup and Installation

```python
# Install required packages
!pip install -q torch torchvision torchaudio
!pip install -q pandas numpy scikit-learn matplotlib seaborn
!pip install -q torch-geometric

# Import libraries
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import Adam
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import mean_squared_error, mean_absolute_error
from torch_geometric.data import HeteroData
from torch_geometric.nn import SAGEConv, to_hetero
from tqdm import tqdm
import time

# Set random seeds for reproducibility
torch.manual_seed(123)
np.random.seed(123)

# Check for GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

# 2. Data Loading and Preprocessing

```python
# Download MovieLens 100K dataset
!wget -nc http://files.grouplens.org/datasets/movielens/ml-100k.zip
!unzip -n ml-100k.zip

# Load data
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=['user_id','movie_id','rating','timestamp'], header=None)
users = pd.read_csv('ml-100k/u.user', sep='|', names=['user_id', 'age', 'gender', 'occupation', 'zip_code'], header=None)
genres = pd.read_csv('ml-100k/u.genre', sep='|', names=['genre', 'genre_id'], header=None)

sm = genres['genre'].tolist()
movies = pd.read_csv('ml-100k/u.item', sep='|',
                     names=['movie_id', 'title', 'release_date', 'video_release_date', 'imdb_url'] + sm,
                     encoding='latin-1', header=None)

print(f"Loaded {len(ratings)} ratings from {len(users)} users and {len(movies)} movies")
print(f"Rating distribution:\n{ratings['rating'].value_counts().sort_index()}")
```

```python
# Preprocess user features
user_features = pd.get_dummies(users, columns=['gender', 'occupation']).drop(['user_id','zip_code'], axis=1)

# Preprocess movie features
movie_features = movies.drop(['title','movie_id','video_release_date','imdb_url'], axis=1)
movie_features['release_date'] = pd.to_datetime(movie_features['release_date'], format='%d-%b-%Y', errors='coerce')
movie_features['release_year'] = movie_features['release_date'].dt.year
movie_features = movie_features.drop(['release_date'], axis=1)
movie_features['release_year'] = movie_features['release_year'].fillna(movie_features['release_year'].mean())

# Normalize age and release year
user_features['age'] = (user_features['age'] - user_features['age'].mean()) / user_features['age'].std()
movie_features['release_year'] = (movie_features['release_year'] - movie_features['release_year'].mean()) / movie_features['release_year'].std()

# Preprocess genre features
genre_features = pd.get_dummies(genres, columns=['genre']).drop(['genre_id'], axis=1)

print(f"User features shape: {user_features.shape}")
print(f"Movie features shape: {movie_features.shape}")
print(f"Genre features shape: {genre_features.shape}")
```

# 3. Graph Construction

```python
# Create heterogeneous graph
data = HeteroData()

# Add node features
data['user'].x = torch.tensor(user_features.astype(float).to_numpy(), dtype=torch.float)
data['movie'].x = torch.tensor(movie_features.astype(float).to_numpy(), dtype=torch.float)
data['genre'].x = torch.tensor(genre_features.astype(float).to_numpy(), dtype=torch.float)

# Prepare edge indices (convert to 0-indexed)
user_ids = ratings['user_id'].to_numpy(dtype=int) - 1
movie_ids = ratings['movie_id'].to_numpy(dtype=int) - 1
rating_values = ratings['rating'].to_numpy(dtype=float)

# User-Movie edges
data['user', 'rates', 'movie'].edge_index = torch.tensor([user_ids, movie_ids], dtype=torch.long)
data['user','rates','movie'].edge_label = torch.tensor(rating_values, dtype=torch.float)

# Reverse edges for bidirectional message passing
data['movie', 'rated_by', 'user'].edge_index = torch.tensor([movie_ids, user_ids], dtype=torch.long)

# Movie-Genre edges
movie_genre_src = []
movie_genre_dst = []
for movie_idx in range(len(movies)):
    movie_row = movies.iloc[movie_idx]
    for genre_id, genre_name in enumerate(sm):
        if movie_row[genre_name] == 1:
            movie_genre_src.append(movie_idx)
            movie_genre_dst.append(genre_id)

data['movie', 'belongs_to', 'genre'].edge_index = torch.tensor([movie_genre_src, movie_genre_dst], dtype=torch.long)
data['genre', 'contains', 'movie'].edge_index = torch.tensor([movie_genre_dst, movie_genre_src], dtype=torch.long)

print("\n" + "="*70)
print("GRAPH CONSTRUCTION COMPLETE")
print("="*70)
print(data)
print("\nNode counts:")
print(f"  Users: {data['user'].x.shape[0]}")
print(f"  Movies: {data['movie'].x.shape[0]}")
print(f"  Genres: {data['genre'].x.shape[0]}")
print("\nEdge counts:")
print(f"  User-Movie ratings: {data['user', 'rates', 'movie'].edge_index.shape[1]}")
print(f"  Movie-Genre: {data['movie', 'belongs_to', 'genre'].edge_index.shape[1]}")
```

# 4. Train/Val/Test Split

```python
# Create 80/10/10 split
num_edges = data['user', 'rates', 'movie'].edge_index.shape[1]
perm = torch.randperm(num_edges)

train_size = int(0.8 * num_edges)
val_size = int(0.1 * num_edges)

train_idx = perm[:train_size]
val_idx = perm[train_size:train_size + val_size]
test_idx = perm[train_size + val_size:]

# Create masks
data['user', 'rates', 'movie'].train_mask = torch.zeros(num_edges, dtype=torch.bool)
data['user', 'rates', 'movie'].val_mask = torch.zeros(num_edges, dtype=torch.bool)
data['user', 'rates', 'movie'].test_mask = torch.zeros(num_edges, dtype=torch.bool)

data['user', 'rates', 'movie'].train_mask[train_idx] = True
data['user', 'rates', 'movie'].val_mask[val_idx] = True
data['user', 'rates', 'movie'].test_mask[test_idx] = True

print(f"Train edges: {train_idx.shape[0]}")
print(f"Val edges: {val_idx.shape[0]}")
print(f"Test edges: {test_idx.shape[0]}")
```

# 5. Model Definition

```python
# Define the base GNN model (homogeneous)
class GNN(torch.nn.Module):
    """Base GNN model that will be converted to heterogeneous."""
    def __init__(self, hidden_channels):
        super().__init__()
        self.conv1 = SAGEConv((-1, -1), hidden_channels, aggr='mean')
        self.conv2 = SAGEConv((-1, -1), hidden_channels, aggr='mean')
        self.dropout = nn.Dropout(0.2)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x


class GraphSAGEModel(torch.nn.Module):
    """Heterogeneous GraphSAGE model for movie rating prediction.

    Architecture:
    - Layer 1: SAGEConv with mean aggregation, hidden_dim=64
    - Activation: ReLU
    - Layer 2: SAGEConv with mean aggregation, hidden_dim=64
    - Prediction: Dot product of user and movie embeddings, scaled to [1,5]
    """
    def __init__(self, hidden_channels, metadata):
        super().__init__()
        # Create base GNN and convert to heterogeneous
        self.gnn = to_hetero(GNN(hidden_channels), metadata, aggr='sum')

    def forward(self, x_dict, edge_index_dict):
        """Forward pass through the model.

        Args:
            x_dict: Dictionary of node features {node_type: features}
            edge_index_dict: Dictionary of edge indices {edge_type: edge_index}

        Returns:
            x_dict: Updated node embeddings
        """
        return self.gnn(x_dict, edge_index_dict)

    def decode(self, z_dict, edge_label_index):
      """Decode edge labels from node embeddings."""
      user_z = z_dict['user'][edge_label_index[0]]
      movie_z = z_dict['movie'][edge_label_index[1]]

      # REVERT TO ORIGINAL
      pred = (user_z * movie_z).sum(dim=-1)
      pred = torch.sigmoid(pred) * 4 + 1  # Scale to [1, 5]

      return pred


# Initialize model
model = GraphSAGEModel(hidden_channels=64, metadata=data.metadata())
model = model.to(device)
data = data.to(device)

print("Model architecture:")
print(model)

# Initialize parameters with a dummy forward pass
with torch.no_grad():
    _ = model(data.x_dict, data.edge_index_dict)

print(f"\nTotal parameters: {sum(p.numel() for p in model.parameters()):,}")
```

# 6. Training Setup

```python
# Hyperparameters
LEARNING_RATE = 0.005
WEIGHT_DECAY = 1e-4
NUM_EPOCHS = 75
BATCH_SIZE = 256

# Optimizer and loss function
optimizer = Adam(model.parameters(), lr=LEARNING_RATE, weight_decay=WEIGHT_DECAY)
criterion = nn.MSELoss()

print("Training Configuration:")
print(f"  Learning rate: {LEARNING_RATE}")
print(f"  Weight decay: {WEIGHT_DECAY}")
print(f"  Batch size: {BATCH_SIZE}")
print(f"  Epochs: {NUM_EPOCHS}")
print(f"  Optimizer: Adam")
print(f"  Loss function: MSELoss")
```

# 7. Training Loop

```python
def evaluate(model, data, mask):
    """Evaluate model on a specific data split.

    Args:
        model: The GraphSAGE model
        data: HeteroData object
        mask: Boolean mask for the split

    Returns:
        mse: Mean squared error
        mae: Mean absolute error
        rmse: Root mean squared error
    """
    model.eval()

    with torch.no_grad():
        # Forward pass
        z_dict = model(data.x_dict, data.edge_index_dict)

        # Get predictions for masked edges
        edge_index = data['user', 'rates', 'movie'].edge_index[:, mask]
        pred = model.decode(z_dict, edge_index)

        # Get ground truth
        target = data['user', 'rates', 'movie'].edge_label[mask]

        # Calculate metrics
        pred_cpu = pred.cpu().numpy()
        target_cpu = target.cpu().numpy()

        mse = mean_squared_error(target_cpu, pred_cpu)
        mae = mean_absolute_error(target_cpu, pred_cpu)
        rmse = np.sqrt(mse)

    return mse, mae, rmse


# Training history
history = {
    'train_loss': [],
    'val_mse': [],
    'val_mae': [],
    'val_rmse': []
}

print("Starting training...\n")
start_time = time.time()

for epoch in range(NUM_EPOCHS):
    model.train()

    # Full-batch training
    optimizer.zero_grad()

    # Forward pass
    z_dict = model(data.x_dict, data.edge_index_dict)

    # Get predictions for training edges
    train_edge_index = data['user', 'rates', 'movie'].edge_index[:, data['user', 'rates', 'movie'].train_mask]
    pred = model.decode(z_dict, train_edge_index)

    # Calculate loss
    target = data['user', 'rates', 'movie'].edge_label[data['user', 'rates', 'movie'].train_mask]
    loss = criterion(pred, target)

    # Backward pass
    loss.backward()
    optimizer.step()

    # Record training loss
    history['train_loss'].append(loss.item())

    # Evaluate on validation set
    val_mse, val_mae, val_rmse = evaluate(model, data, data['user', 'rates', 'movie'].val_mask)
    history['val_mse'].append(val_mse)
    history['val_mae'].append(val_mae)
    history['val_rmse'].append(val_rmse)

    # Print progress
    if (epoch + 1) % 5 == 0 or epoch == 0:
        print(f"Epoch {epoch+1:3d}/{NUM_EPOCHS} | "
              f"Train Loss: {loss.item():.4f} | "
              f"Val RMSE: {val_rmse:.4f} | "
              f"Val MAE: {val_mae:.4f}")

training_time = time.time() - start_time
print(f"\nTraining completed in {training_time:.2f} seconds ({training_time/60:.2f} minutes)")
```

# 8. Final Evaluation on Test Set

```python
# Evaluate on test set
test_mse, test_mae, test_rmse = evaluate(model, data, data['user', 'rates', 'movie'].test_mask)

print("="*70)
print("FINAL TEST SET RESULTS")
print("="*70)
print(f"Test MSE:  {test_mse:.4f}")
print(f"Test RMSE: {test_rmse:.4f}")
print(f"Test MAE:  {test_mae:.4f}")
print("="*70)
```

# 10. Baseline Comparison

```python
# Simple baselines for comparison

# Baseline 1: Global mean
train_ratings = data['user', 'rates', 'movie'].edge_label[data['user', 'rates', 'movie'].train_mask]
global_mean = train_ratings.mean().item()
test_target_np = test_target.cpu().numpy()

baseline_pred = np.full_like(test_target_np, global_mean)
baseline_mse = mean_squared_error(test_target_np, baseline_pred)
baseline_rmse = np.sqrt(baseline_mse)
baseline_mae = mean_absolute_error(test_target_np, baseline_pred)

# Baseline 2: User mean
user_means = {}
train_edge_index = data['user', 'rates', 'movie'].edge_index[:, data['user', 'rates', 'movie'].train_mask]
for i in range(len(train_edge_index[0])):
    user_id = train_edge_index[0, i].item()
    rating = train_ratings[i].item()
    if user_id not in user_means:
        user_means[user_id] = []
    user_means[user_id].append(rating)

user_means = {k: np.mean(v) for k, v in user_means.items()}

user_baseline_pred = []
for i in range(len(test_edge_index[0])):
    user_id = test_edge_index[0, i].item()
    user_baseline_pred.append(user_means.get(user_id, global_mean))

user_baseline_pred = np.array(user_baseline_pred)
user_baseline_mse = mean_squared_error(test_target_np, user_baseline_pred)
user_baseline_rmse = np.sqrt(user_baseline_mse)
user_baseline_mae = mean_absolute_error(test_target_np, user_baseline_pred)

# Comparison table
comparison_df = pd.DataFrame({
    'Model': ['Global Mean', 'User Mean', 'GraphSAGE (Ours)'],
    'MSE': [f"{baseline_mse:.4f}", f"{user_baseline_mse:.4f}", f"{test_mse:.4f}"],
    'RMSE': [f"{baseline_rmse:.4f}", f"{user_baseline_rmse:.4f}", f"{test_rmse:.4f}"],
    'MAE': [f"{baseline_mae:.4f}", f"{user_baseline_mae:.4f}", f"{test_mae:.4f}"]
})

print("\n" + "="*70)
print("BASELINE COMPARISON")
print("="*70)
print(comparison_df.to_string(index=False))
print("="*70)

# Calculate improvement
improvement_over_global = ((baseline_rmse - test_rmse) / baseline_rmse) * 100
improvement_over_user = ((user_baseline_rmse - test_rmse) / user_baseline_rmse) * 100

print(f"\nImprovement over Global Mean: {improvement_over_global:.2f}%")
print(f"Improvement over User Mean: {improvement_over_user:.2f}%")

comparison_df.to_csv('baseline_comparison.csv', index=False)
print("\nComparison table saved as 'baseline_comparison.csv'")
```
